```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
#read CPI dataset and abstract useful data
cpi_init = pd.read_csv('Data/CPI.Food')
lst_y, lst_m, value = [], [], []
for i in range(len(cpi_init)):
    if cpi_init.iloc[i, 0].split('\t')[0].strip() == 'APU0000701111':
        lst_y.append(cpi_init.iloc[i, 0].split('\t')[1].strip())
        lst_m.append(cpi_init.iloc[i, 0].split('\t')[2].strip())
        value.append(cpi_init.iloc[i, 0].split('\t')[3].strip())
```

```python
#change the useful data into dataframe
cpi = pd.DataFrame(columns=['year', 'month', 'value'])
cpi['year'] = lst_y
cpi['month'] = lst_m
cpi['value'] = value
cpi['value'] = cpi['value'].astype(float)
cpi['year'] = cpi['year'].astype(int)
cpi['month'] = [i[1:] for i in cpi['month']]
cpi = cpi.loc[cpi['year'] >= 2000]
cpi = cpi.loc[cpi['year'] <= 2017]
cpi = cpi.reset_index(drop = True)
new_cpi = pd.DataFrame(columns = ['year', 'quarter', 'value'])
new_lst_y,new_lst_q,new_lst_value = [],[],[]
count = 0
total = 0
for i in cpi['value']:
    count+=1
    if count == 3:
        total += i
        new_lst_value.append(total)
        total = 0
        count = 0
    else:
        total += i
for i in range(2000,2018):
    for j in range(1,5):
        new_lst_y.append(i)
        new_lst_q.append(j)
new_cpi['year'] = new_lst_y
new_cpi['quarter'] = new_lst_q
new_cpi['value'] = new_lst_value
new_cpi.to_csv('cpi.csv')
```

```python
#read GNI dataset
gni_init = pd.read_csv('Data/GNI.csv')
lst_y = []
value = []
for i in range((2017-1970)):
    lst_y.append((1970+i+1))
    value.append(gni_init.iloc[0,(i+5)])
gni = pd.DataFrame(columns=['year','value'])
gni['year'] = lst_y
gni['value'] = value
gni['year'] = gni['year'].astype(int)
gni['value'] = gni['value'].astype(float)
gni = gni.loc[gni['year'] >= 2000]
gni = gni.reset_index(drop = True)
new_lst_value = []
for i in gni['value']:
    for _ in range(4):
        new_lst_value.append(i/4)
new_gni = pd.DataFrame(columns = ['year', 'quarter', 'value'])
new_gni['year'] = new_lst_y
new_gni['quarter'] = new_lst_q
new_gni['value'] = new_lst_value
new_gni.to_csv('gni.csv')
```

```python
#read livestocks price dataset
price_init = pd.read_excel('Data/LivestockPrices.xlsx',sheet_name='historical')
price_init = price_init.drop(columns = ['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4', \
                                        'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', \
                                        'Unnamed: 9', 'Unnamed: 11', 'Unnamed: 12', \
                                        'Unnamed: 13'])
lst_t = []
lst_b = []
lst_p = []
lst_l = []
for i in range(3,len(price_init)):
    lst_t.append(price_init.iloc[i, 0])
    lst_b.append(price_init.iloc[i, 1])
    lst_p.append(price_init.iloc[i, 2])
    lst_l.append(price_init.iloc[i, 3])
price_df = pd.DataFrame(columns = ['year','beef_price','pork_price','lamb_price','quarter'])
price_df['year'] = lst_t
price_df['beef_price'] = lst_b
price_df['pork_price'] = lst_p
price_df['lamb_price'] = lst_l
price_df['year'] = price_df['year'].astype(str)
price_df['month'] = [i[5:7] for i in price_df['year']]
price_df['month'] = price_df['month'].astype(str)
price_df['year'] = [i[:4] for i in price_df['year']]
price_df['year'] = price_df['year'].astype(int)
price_df['beef_price'] = price_df['beef_price'].astype(float)
price_df['pork_price'] = price_df['pork_price'].astype(float)
price_df['lamb_price'] = price_df['lamb_price'].astype(float)
price_df = price_df.loc[price_df['year'] <= 2017]
price_df = price_df.reset_index(drop=True)
lst_q = []
count = 0
for i in range(18):
    for j in range(1,5):
        for _ in range(3):
            lst_q.append(j)
price_df['quarter'] = lst_q
price_df.groupby(['year','quarter']).mean().reset_index().to_csv('price.csv')
```

```python
#read MeatSD dataset
def read_data_to_df(sheet_name):
    beef_stock_init = pd.read_excel('Data/MeatSDFull.xlsx',sheet_name=sheet_name)
    beef_stock_init = beef_stock_init.fillna(method='ffill')
    lst_y, lst_m = [],[]
    lst_commercial, lst_farm, lst_beginstocks, lst_imports, lst_exports, lst_endstocks, lst_pop = [],[],[],[],[],[],[]
    beef_stock_init.head()
    for i in range(2,173):
        lst_y.append(beef_stock_init.iloc[i,0])
        lst_m.append(beef_stock_init.iloc[i,1])
        lst_commercial.append(beef_stock_init.iloc[i,2])
        lst_farm.append(beef_stock_init.iloc[i,3])
        lst_beginstocks.append(beef_stock_init.iloc[i,5])
        lst_imports.append(beef_stock_init.iloc[i,6])
        lst_exports.append(beef_stock_init.iloc[i,8])
        lst_endstocks.append(beef_stock_init.iloc[i,9])
        lst_pop.append(beef_stock_init.iloc[i,11])
    beef_stock = pd.DataFrame(columns=['year','month','commercial','farm', \
                                      'begin_stocks','end_stocks','imports', \
                                      'exports','population'])
    beef_stock['year'] = lst_y
    beef_stock['month'] = lst_m
    beef_stock['commercial'] = lst_commercial
    beef_stock['farm'] = lst_farm
    beef_stock['begin_stocks'] = lst_beginstocks
    beef_stock['end_stocks'] = lst_endstocks
    beef_stock['imports'] = lst_imports
    beef_stock['exports'] = lst_exports
    beef_stock['population'] = lst_pop
    beef_stock['year'] = beef_stock['year'].astype(int)
    beef_stock['commercial'] = beef_stock['commercial'].astype(float)
    beef_stock['farm'] = beef_stock['farm'].astype(float)
    beef_stock['begin_stocks'] = beef_stock['begin_stocks'].astype(float)
    beef_stock['end_stocks'] = beef_stock['end_stocks'].astype(float)
    beef_stock['imports'] = beef_stock['imports'].astype(float)
    beef_stock['exports'] = beef_stock['exports'].astype(float)
    beef_stock['population'] = beef_stock['population'].astype(float)
    beef_stock = beef_stock.loc[beef_stock['month'] != 'Yr Jan-Dec']
    beef_stock = beef_stock.loc[beef_stock['year']>= 2000]
    beef_stock = beef_stock.loc[beef_stock['year']<= 2017]
    beef_stock = beef_stock.reset_index(drop = True)
    return beef_stock
```

```python
#beef stocks
beef_stock_df = read_data_to_df('WASDE_Beef-Full')
beef_stock_df['month'] = [int(i[1]) for i in beef_stock_df['month']]
beef_stock_df.to_csv('beef_stock.csv')
```

```python
#pork stocks
pork_stock_df = read_data_to_df('WASDE_Pork-Full')
pork_stock_df['month'] = [int(i[1]) for i in pork_stock_df['month']]
pork_stock_df.to_csv('pork_stock.csv')
```

```python
#lamb stocks
lamb_stock_df = read_data_to_df('WASDE_LambMutton-Full')
lamb_stock_df['month'] = [int(i[1]) for i in lamb_stock_df['month']]
lamb_stock_df.to_csv('lamb_stock.csv')
```
